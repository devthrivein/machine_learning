```python
from PIL import Image
import os
```

```python
def image_resize(image_path, dir_path, target_size):
    image = Image.open(os.path.join(dir_path, image_path))
    resized_image = image.resize(target_size)
    resized_image.save(os.path.join(dir_path, image_path))
```

```python
def main(dir_path, target_size):
    image_list = os.listdir(dir_path)

    counter = 0
    print(f'Found {len(image_list)} images')
    print(f'Resizing...')
    for x, item in enumerate(image_list):
        image_resize(item, dir_path, target_size)
        counter +=1

    print(f'Successfully resized {counter} images')
```

```python
dir_path = r'C:\\Users\\brill\\Jupyter\\Bangkit2023\\Capstone\\Dataset\\data\\'
target_size = (500, 500)

main(dir_path, target_size)
```
